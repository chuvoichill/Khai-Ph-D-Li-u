```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
df=pd.read_csv("car_price_prediction.csv")
df
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

# Xử lý dữ liệu 

```python
X=pd.get_dummies(X,dtype=int)
X.head(10)
```

# Thực hành vời dữ liệu chưa chuẩn hóa

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X["Cylinders"]=scaler.fit_transform(X[["Cylinders"]])
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=4)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# KNN regresstion

```python
knn_Reg=KNeighborsRegressor(n_neighbors=9)
knn_Reg.fit(X_train,y_train)
y_pred_knn=knn_Reg.predict(X_test)
print(pd.DataFrame({"y_test":y_test,"y_predicted":y_pred_knn}).head(10))
```

# Đánh giá

```python
from sklearn.metrics import mean_absolute_error
MAE_KNN=mean_absolute_error(y_test,y_pred_knn)
print("MAE:",MAE_KNN)
```

# Linear regresstion

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit([[1],[2],[3],[4],[5],[6]],[11,11,14,16,18,20])
print(reg.coef_)
print(reg.intercept_)
```
