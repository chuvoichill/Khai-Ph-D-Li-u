```python
from sklearn.preprocessing import MinMaxScaler
df=pd.read_excel('Bank_full.xlsx')
df
```

```python
scaler = MinMaxScaler(feature_range=(1,5))
model=scaler.fit(df[["balance"]])
df["balance"]=model.transform(df[["balance"]])
df
```

```python
#Quay lại giá trị cũ
inverse = model.inverse_transform(df[["balance"]])
print(inverse)
```

```python
#StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
model=scaler.fit(df[["balance"]])
df["balance"]=model.transform(df[["balance"]])
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 8))
plt.bar(x, y)
plt.plot(x, y, marker='o', markersize=10, linestyle='-.', linewidth=2, color='red')
plt.xlabel('Job', fontsize=16)
plt.ylabel('People', fontsize=16)
```

```python
plt.figure(figsize=(80, 40))
plt.pie(y,
        labels = x, # Nhãn của các nhóm
        colors = ['red', 'blue', 'green'], # Màu sắc của các nhóm
        autopct = '%1.1f%%', # Format hiển thị giá trị %
        shadow = False,
        textprops={'fontsize': 50})
plt.title("JOB", fontsize=80)
```
